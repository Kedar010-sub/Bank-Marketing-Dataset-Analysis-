```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/kaggle/input/imdb-data-set/IMDB-Movie-Data.csv')
```

```python
data.head
```

```python
data.shape
```

To get information about dataset

```python
#to get information about dataset
data.info()
```

Checking missing values in the dataset

```python
#check missing values in the dataset
data.isnull().sum()
```

```python
sns.heatmap(data.isnull())
plt.show()
```

Map For Checking Missing Values 

```python
#to get missing values percentage 
percentage_missing=data.isnull().sum()*100/len(data)
print(percentage_missing)
```

```python
#drop all missing values
data.dropna(axis=0)
```

```python
#check for duplicate data
duplicate_data=data.duplicated().any()
print(duplicate_data)
```

```python
#to get statistics about the dataframe
data.describe(include='all') 
```

```python
#display title of the movie  having runtime >=180minutes
data.columns
```

```python
data[data['Runtime (Minutes)']>=180]['Title']
```

```python
#in which year there was the highest avarage voting
data.groupby('Year')['Votes'].mean().sort_values(ascending=False)
```

```python
sns.barplot(x='Year',y='Votes',data=data)
plt.title('Avarage votes by year')
plt.show()
```

Average Votes By years 

```python
#in which year there was the highest avarage revenue 
data.columns
```

```python
data.groupby('Year')['Revenue (Millions)'].mean().sort_values(ascending=False)
```

```python
sns.barplot(x='Year',y='Revenue (Millions)',data=data)
plt.title('Highest Avarage Revenue votes by year')
plt.show()
```

Highest Average revenue votes by year 

```python
#to get the average rating for each director 
data.columns
```

```python
data.groupby('Director')['Rating'].mean().sort_values(ascending= False)
```

```python
#To display top 10 Lenghty movie titles and runtime
data.columns
```

```python
top10=data.nlargest(10,'Runtime (Minutes)')[['Title','Runtime (Minutes)']]\
.set_index('Title')
print(top10)
```

```python
sns. barplot(x='Runtime (Minutes)',y=top10.index,data=top10)
plt.show()
```
